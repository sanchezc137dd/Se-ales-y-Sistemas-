```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 10)
x = np.where(n < 4, 1, 0)

plt.stem(n, x)
plt.title("Señal x[n] = 1 - Σ δ[n - 1 - k], k=3..∞")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el eje temporal
t = np.linspace(-5, 5, 1000)

# Aproximación de delta de Dirac como picos estrechos
def delta(t, t0, width=0.01):
    return np.where(np.abs(t - t0) < width, 1/width, 0)

# Señal x(t): combinación de dos deltas
x = delta(t, -2) - delta(t, 2)

# Señal y(t): integral acumulativa de x(t)
dt = t[1] - t[0]
y = np.cumsum(x) * dt

# Graficar
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(t, x, label='x(t) = δ(t + 2) - δ(t - 2)')
plt.title('Señal x(t)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, y, label='y(t) = ∫ x(τ) dτ', color='orange')
plt.title('Señal y(t) acumulada')
plt.xlabel('Tiempo t')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el eje temporal
t = np.linspace(-4, 4, 1000)

# Definir la señal por partes
x = np.piecewise(t,
                 [np.abs(t) <= 1,
                  (np.abs(t) > 1) & (np.abs(t) <= 2),
                  np.abs(t) > 2],
                 [lambda t: t**2,
                  1,
                  lambda t: np.abs(t) - 2])

# Graficar
plt.figure(figsize=(8, 4))
plt.plot(t, x, label='x(t)', color='purple')
plt.title('Señal por partes definida en función de |t|')
plt.xlabel('Tiempo t')
plt.ylabel('Amplitud x(t)')
plt.grid(True)
plt.legend()
plt.show()
```
